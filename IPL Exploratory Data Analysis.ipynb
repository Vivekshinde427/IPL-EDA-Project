## ü™ô Toss Winners Analysis

The following table shows which teams have won the most tosses across all IPL matches.  
Toss decisions can influence match outcomes, especially when pitch or weather conditions matter.

Let's look at the top teams by number of toss wins.


```python
df['toss_winner'].value_counts()
```

```python
# Create columns
df['toss_match_win'] = df['toss_winner'] == df['winner']

# Grouped data
toss_result = df.groupby('toss_winner')['toss_match_win'].value_counts().unstack().fillna(0)
toss_result.columns = ['Lost Match After Toss', 'Won Match After Toss']
toss_result = toss_result.sort_values(by='Won Match After Toss', ascending=False)

# Plot
toss_result.plot(kind='barh', stacked=True, figsize=(10,7), colormap='Accent')
plt.title('üèè Toss Outcomes: Match Win vs Loss by Toss Winner')
plt.xlabel('Number of Matches')
plt.ylabel('Teams')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

### üèè Most Playoff Appearances by IPL Teams

In IPL, the playoffs usually consist of the last 4 matches of each season (Qualifier 1, Eliminator, Qualifier 2, and Final).  
Here, we identify these matches by selecting the top 4 matches by `id` for each season and count how many times each team appeared in those playoff matches.

We consider both teams playing in each match (`team1` and `team2`) to calculate the total playoff appearances for every team.

The following code aggregates the playoff matches across all seasons and computes the number of playoff appearances by each team.


```python
playoff_matches_per_season = 4

playoffs_list = []

for season in df['Season'].unique():
    season_matches = df[df['Season'] == season]
    playoffs = season_matches.sort_values('id', ascending=False).head(playoff_matches_per_season)
    playoffs_list.append(playoffs)


playoffs_df = pd.concat(playoffs_list)


team_playoff_counts = pd.concat([playoffs_df['team1'], playoffs_df['team2']]).value_counts()

print("Most Playoff Appearances by Teams:")
print(team_playoff_counts)
```

<img src="ipl csk.avif" width="100%" />


```python
team_colors = {
    'Mumbai Indians': 'blue',
    'Chennai Super Kings': 'yellow',
    'Royal Challengers Bangalore': 'red',
    'Kolkata Knight Riders': 'purple',
    'Sunrisers Hyderabad': 'orange',
    'Delhi Capitals': 'skyblue',
    'Rajasthan Royals': 'pink',
    'Punjab Kings': 'green',
}

team_playoff_counts_sorted = team_playoff_counts.sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(
    x=team_playoff_counts_sorted.index, 
    y=team_playoff_counts_sorted.values,
    palette=[team_colors.get(team, 'grey') for team in team_playoff_counts_sorted.index]
)
plt.title("üèÜ Most Playoff Appearances by IPL Teams")
plt.xlabel("Teams")
plt.ylabel("Number of Playoff Appearances")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
